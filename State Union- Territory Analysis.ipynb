```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data for the analysis (the same as above, plus crime statistics)
data = {
    "State": ["Andhra Pradesh", "Arunachal Pradesh", "Assam", "Bihar", "Chhattisgarh", 
              "Goa", "Gujarat", "Haryana", "Himachal Pradesh", "Jammu & Kashmir",
              "Jharkhand", "Karnataka", "Kerala", "Madhya Pradesh", "Maharashtra",
              "Manipur", "Meghalaya", "Mizoram", "Nagaland", "Odisha",
              "Punjab", "Rajasthan", "Sikkim", "Tamil Nadu", "Tripura",
              "Uttar Pradesh", "Uttarakhand", "West Bengal"],
    
    "Population_2011": [84665533, 1382611, 31169272, 104099452, 25545198,
                        1458545, 60439692, 25353081, 6864602, 12541302,
                        32988134, 61095297, 33406061, 72626809, 112374333,
                        2721756, 2964007, 1097206, 1978502, 41947358,
                        27743338, 68548437, 610577, 72147030, 3673917,
                        199812341, 10086292, 91276115],
    
    "Literacy_Rate_2011": [67.02, 65.38, 72.19, 61.8, 70.28,
                           87.40, 78.03, 76.64, 82.80, 67.16,
                           66.41, 75.36, 94.00, 69.32, 82.34,
                           79.21, 74.43, 91.33, 79.55, 72.87,
                           75.84, 66.11, 81.42, 80.09, 87.22,
                           69.72, 78.82, 77.08],
    
    "Area_km2": [162968, 83743, 78438, 94163, 135191,
                 3702, 196024, 44212, 55673, 222236,
                 79714, 191791, 38863, 308350, 307713,
                 22327, 22429, 21081, 16579, 155707,
                 50362, 342239, 7096, 130058, 10486,
                 243286, 53483, 88752],

    # Hypothetical crime data (Total crimes reported in 2011)
    "Total_Crimes_2011": [212500, 3000, 68000, 150000, 55000,
                          5000, 120000, 85000, 12000, 35000,
                          60000, 140000, 8000, 190000, 250000,
                          4000, 5000, 3000, 2000, 70000,
                          40000, 180000, 1000, 160000, 5000,
                          250000, 20000, 190000]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv("india_state_crime_data_2011.csv", index=False)

# Display the DataFrame
df

# Analysis of Literacy Rate vs Total Crimes
plt.figure(figsize=(14, 8))
sns.scatterplot(x='Literacy_Rate_2011', y='Total_Crimes_2011', data=df)
plt.title('Literacy Rate vs Total Crimes (2011)')
plt.xlabel('Literacy Rate (%)')
plt.ylabel('Total Crimes')
plt.grid(True)
plt.show()

# Analysis of Area vs Overall Crime
plt.figure(figsize=(14, 8))
sns.scatterplot(x='Area_km2', y='Total_Crimes_2011', data=df)
plt.title('Area vs Total Crimes (2011)')
plt.xlabel('Area (km^2)')
plt.ylabel('Total Crimes')
plt.grid(True)
plt.show()

# Analysis of Population vs Overall Crime
plt.figure(figsize=(14, 8))
sns.scatterplot(x='Population_2011', y='Total_Crimes_2011', data=df)
plt.title('Population vs Total Crimes (2011)')
plt.xlabel('Population')
plt.ylabel('Total Crimes')
plt.grid(True)
plt.show()

# Crime rate per 100,000 people
df['Crime_Rate_per_100000'] = (df['Total_Crimes_2011'] / df['Population_2011']) * 100000

# Analysis of Crime Rate per 100,000 people vs Literacy Rate
plt.figure(figsize=(14, 8))
sns.scatterplot(x='Literacy_Rate_2011', y='Crime_Rate_per_100000', data=df)
plt.title('Literacy Rate vs Crime Rate per 100,000 people (2011)')
plt.xlabel('Literacy Rate (%)')
plt.ylabel('Crime Rate per 100,000 people')
plt.grid(True)
plt.show()

# Year-on-Year total crime rate analysis
# Hypothetical yearly data
years = list(range(2001, 2013))
total_crimes_yearly = [1500000, 1520000, 1580000, 1600000, 1650000, 1700000, 1750000, 1800000, 1850000, 1900000, 1950000, 2000000]

plt.figure(figsize=(14, 8))
plt.plot(years, total_crimes_yearly, marker='o')
plt.title('Year-on-Year Total Crime Rate (2001-2012)')
plt.xlabel('Year')
plt.ylabel('Total Crimes')
plt.grid(True)
plt.show()

# Each state crime report
for index, row in df.iterrows():
    print(f"State: {row['State']}")
    print(f"  Population (2011): {row['Population_2011']}")
    print(f"  Literacy Rate (2011): {row['Literacy_Rate_2011']}%")
    print(f"  Area: {row['Area_km2']} km^2")
    print(f"  Total Crimes (2011): {row['Total_Crimes_2011']}")
    print(f"  Crime Rate per 100,000 people: {row['Crime_Rate_per_100000']:.2f}")
    print("\n")

```
